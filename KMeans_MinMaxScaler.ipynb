```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km = KMeans(n_clusters=3)   #Created K means object having 3 clusters
```

```python
y_predicted = km.fit_predict(df.drop('Name',axis='columns'))
```

```python
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='black')
plt.xlabel('Age')
plt.ylabel('Income($)')
```

### Scaling is not proper so there is mis classification - We need to preprocess for Scaling - MinMaxScaler

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])   #fits / normalizes from 0 to 1
df['Income($)'] = scaler.transform(df['Income($)'].values.reshape(-1,1))

scaler.fit(df[['Age']])
df.Age = scaler.transform(df.Age.values.reshape(-1,1))
df
```

```python
km2 = KMeans(n_clusters=3)
y_pred = km2.fit_predict(df.drop('Name',axis='columns'))
y_pred
```

```python
df['cluster'] = y_pred
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='red')
plt.scatter(df3.Age,df3['Income($)'],color='black')
plt.xlabel('Age')
plt.ylabel('Income($)')
```
