## Gradient Descent and Mean Squared Error(Cost Function)

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y= np.array([5,7,9,11,13])
```

### Partial Derivative Equations

### Cost Function - MSE

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations =  1000
    n = len(x)
    learning_rate = 0.001
    for i in range(iterations):
        y_pred = m_curr * x + b_curr

        cost = (1/n) * sum([val**2 for val in (y-y_pred)])
        
        m_derivative = -(2/n)*sum(x*(y-y_pred))   #partial derivative
        b_derivative = -(2/n)*sum(y-y_pred)
        m_curr = m_curr - learning_rate * m_derivative   #adjusting the slope using gradient descent
        b_curr = b_curr - learning_rate * b_derivative   #adjusting the intercept by gradient descent for best fit line
        if i % 10 == 9:
            print(f"m: {m_curr}, b: {b_curr}, Iteration_no : {i}, Cost : {cost}")
        
```

```python
gradient_descent(x,y)
```
